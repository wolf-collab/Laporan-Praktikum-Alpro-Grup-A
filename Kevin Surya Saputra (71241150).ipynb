```python
#Latihan Mandiri 1.1
%matplotlib inline

import matplotlib.pyplot as plt
import numpy as np

x = np.linspace(0, 10)
y = np.sin(x)
z = np.cos(x)

plt.plot(x, y, 'b', x,z, 'r')
plt.xlabel('Radians')
plt.ylabel('Value')
plt.title('Plotting Demonstration')
plt.legend(['Sin', 'Cos'])
plt.grid()
plt.show()
```

```python
#Latihan Mandiri 1.2
#Keadaan mula-mula
print('Tahap 1:')
harga_beli_awal = 650000
emasnya_Gerard_awal = 25
total_harga_beli_awal = harga_beli_awal * emasnya_Gerard_awal
print('1. Modal Awal yang Gerard keluarkan = Rp. ',total_harga_beli_awal)

#Harga emas naik pertama jadi Rp. 685.000
print('\nTahap 2:')
harga_naik_pertama = 685000
harga_jual_emas_pertama = harga_naik_pertama * emasnya_Gerard_awal
print('1. Harga emas yang Gerard beli menjadi = Rp. ',harga_jual_emas_pertama)

#Keuntungan yang didapatkan Gerard
cuan = harga_jual_emas_pertama - total_harga_beli_awal
print('2. Keuntungan yang didapatkan Gerard (dalam Rp)= Rp. ',cuan)

#keuntungan dalam persen
persen = round((cuan / total_harga_beli_awal) * 100, 2)
print('3. Keuntungan yang didapatkan Gerard (dalam %)= ',persen, '%')

#Beli lagi emas 15 gram
print('\nTahap 3:')
beli_emas_sebanyak = 15
dengan_harga = 685000
total_harga = beli_emas_sebanyak * dengan_harga
print('1. Harga emas yang Gerard bayar = Rp. ',total_harga)

#Harga Emas naik lagi jadi Rp 715.000
harga_naik_kedua = 715000
banyak_emas_Gerard = emasnya_Gerard_awal + beli_emas_sebanyak
print('2. Total Berat emasnya Gerard (Gram) = ',banyak_emas_Gerard,'Gram')
total_emasnya_Gerard =  total_harga_beli_awal + total_harga
print('3. Total Harga Emas Yang Gerard Beli = Rp. ',total_emasnya_Gerard)

# Total harga emas dan dijual
total_harga_jual = banyak_emas_Gerard * harga_naik_kedua
print('4. Total Harga Jual Emas Terbaru = Rp. ',total_harga_jual)
Total_keuntungan = total_harga_jual - total_emasnya_Gerard
print('5. Keuntungan Yang Didapatkan Gerard (dalam Rp)= Rp. ',Total_keuntungan)
total_persen = round((Total_keuntungan / total_emasnya_Gerard) * 100, 2)
print('6. Keuntungan Yang Didapatkan Gerard (dalam %)= ',total_persen, '%')
```

```python
dana_awal = 200000000
target_saldo = 400000000
bunga = 0.1
tahun = 0
saldo = dana_awal

#perulangan untuk meddapat 400.000.000
while saldo < target_saldo:
    tahun += 1
    saldo *= (1 + bunga)
    print(f'Tahun {tahun}: Saldo = Rp. {saldo:,.2f}')

#Hasil Akhir
print(f'\nWaktu yang diperlukan: {tahun} tahun.')
print(f'Saldo akhir: Rp {saldo:,.2f}')
```
